```python
# QA 완료
import pandas as pd
import numpy as np
import time
import math

####################################################Teams,Holdings,Subsidiarys.csv 파일 위치 >>>
Teams_df = pd.read_csv('../V2/data/Teams.csv.csv')
Holdings_df = pd.read_csv('../V2/data/Holdings.csv.csv')
Subsidiarys_df = pd.read_csv('../V2/data/Subsidiarys.csv.csv')

# 팀즈, 홀딩즈 스프레드 시트
############################################# !! 구글 드라이브 내에 다음 이름과 중복된 이름의 파일이 있으면 오류납니다 !!
# Teams_GSC0 = BPCC.open("Teams")
# Teams_GSC = Teams_GSC0.sheet1

# Holdings_GSC0 = BPCC.open("Holdings")
# Holdings_GSC = Holdings_GSC0.sheet1

# Subsidiarys_GSC0 = BPCC.open("Subsidiarys")
# Subsidiarys_GSC = Subsidiarys_GSC0.sheet1

Teams_df.set_index('Team', inplace=True)
Holdings_df.set_index('Team', inplace=True)
Subsidiarys_df.set_index('Team', inplace=True)
Teams_df['subsidiary'] = [list() for _ in range(len(Teams_df))]
#백업 DF
Teams_dict = {}
for w in range(1, 7):
    Teams_dict[f'Teams{w}_df'] = Teams_df.copy()
Holdings_dict = {}
for w in range(1, 7):
    Holdings_dict[f'Holdings{w}_df'] = Holdings_df.copy()
Subsidiarys_dict = {}
for w in range(1, 7):
    Subsidiarys_dict[f'Subsidiarys{w}_df'] = Subsidiarys_df.copy()
LastOrderList= list()

print(Teams_df)
print(Holdings_df)
print(Subsidiarys_df)

TradeOrder_df = pd.read_csv('../V2/data/BP_TradeOrder.csv')
print(TradeOrder_df.head())
```

```python
# QA 완료
def TeamsReload():
    Teams_df.reset_index(drop=False, inplace=True)
    Teams_df['subsidiary'] = Teams_df['subsidiary'].apply(lambda x: ','.join(map(str, x)) if isinstance(x, list) else str(x))
    Teams_df['capital'] = Teams_df['capital'].map('{:,}'.format)
    Teams_df['price'] = Teams_df['price'].map('{:,}'.format)
    Teams_df['total asset'] = Teams_df['total asset'].map('{:,}'.format)
    TeamsList = [Teams_df.columns.values.tolist()] + Teams_df.values.tolist()
    Teams_df.set_index('Team', inplace=True)
    Teams_df["capital"] = Teams_df["capital"].str.replace(",", "", regex=False).astype(float).astype(int)
    Teams_df["price"] = Teams_df["price"].str.replace(",", "", regex=False).astype(float).astype(int)
    Teams_df["total asset"] = Teams_df["total asset"].str.replace(",", "", regex=False).astype(float).astype(int)
    Teams_df['subsidiary'] = Teams_df['subsidiary'].astype(object)
    for i in Teams_df.index:
        val = Teams_df.at[i, 'subsidiary']
        if val == '':
            Teams_df.at[i, 'subsidiary'] = []
        else:
            Teams_df.at[i, 'subsidiary'] = val.split(',')
    Teams_df.to_csv('../V2/data/Teams.csv.csv')
def HoldingsReload():
    Holdings_df.reset_index(drop=False, inplace=True)
    HoldingsList = [Holdings_df.columns.values.tolist()] + Holdings_df.values.tolist()
    Holdings_df.set_index('Team', inplace=True)
    Holdings_df.to_csv('../V2/data/Holdings.csv.csv')
def SubsidiarysReload():
    Subsidiarys_df.reset_index(drop=False, inplace=True)
    SubsidiarysList = [Subsidiarys_df.columns.values.tolist()] + Subsidiarys_df.values.tolist()
    Subsidiarys_df.set_index('Team', inplace=True)
    Subsidiarys_df.to_csv('../V2/data/Subsidiarys.csv.csv')
TeamsReload()
HoldingsReload()
SubsidiarysReload()
```

```python
# QA 완료
def RunMinigame():
    global roundnum
    while True:
        try:roundnum = int(input('현재 라운드 수를 입력해주세요 : '))
        except ValueError:
            print('숫자만 입력하십시오')
            continue
        break
    print(f'\n{roundnum}차 미니게임 시작!\n')
    while True:
        Winners = input('미니게임 등수 순서대로 입력해주세요. (EX: A,D,E): ')
        WinList = [w.strip() for w in Winners.split(',')]

        # 리스트 유효성 검사
        if (
            isinstance(WinList, list) and
            len(WinList) == 3 and
            len(set(WinList)) == 3 and
            set(WinList).issubset({'A','B','C','D','E','F','G','H'})
        ):
            print("입력 성공:", WinList)
            break
        else:
            print("형식이 잘못됐습니다. 다시 입력해주세요.")

    FirstTeam = WinList[0]
    SecondTeam = WinList[1]
    ThirdTeam = WinList[2]
    AllTeam = ['A', 'B', 'C', 'D', 'E', 'F', 'G','H']
    RestTeam = list(set(AllTeam)-set(WinList))

    Teams_df.loc[FirstTeam,'capital'] = Teams_df.loc[FirstTeam,'capital'] + round(300000*((1.3)**(roundnum-1)),-2)
    Teams_df.loc[SecondTeam,'capital'] = Teams_df.loc[SecondTeam,'capital'] + round(200000*((1.3)**(roundnum-1)),-2)
    Teams_df.loc[ThirdTeam,'capital'] = Teams_df.loc[ThirdTeam,'capital'] + round(100000*((1.3)**(roundnum-1)),-2)
    for t in RestTeam:
        Teams_df.loc[t,'capital'] = Teams_df.loc[t,'capital'] + round(50000*((1.3)**(roundnum-1)),-2)

    IsSubsidiary()

    #자회사 하나마다 수익 20만원
    for w in Teams_df.drop(index=['S']).index:
        if len(Teams_df.at[w,'subsidiary']) != 0:
            baedang = 200000*len(Teams_df.at[w,'subsidiary'])
            Teams_df.at[w,'capital'] = Teams_df.at[w,'capital'] + baedang
            print(f'{Teams_df.at[w,'team name']}에게 자회사 보유 수익금 {baedang}원이 지급되었습니다!')
    Winner2()
    MoneyMoney()
```

```python
#QA 완료
def ForcedTrade():
# 자회사 강매권 사용
    for w in Teams_df.drop(index=['S']).index:
        if Teams_df.at[w,'parent'] != 'X':
            while True:
                  SubsidiaryAnswer=input(f'자회사{Teams_df.at[w,'team name']}팀의 강매 여부를 입력해주세요 O/X : ')
                  if SubsidiaryAnswer == 'X':
                      print(f'자회사{Teams_df.at[w,'team name']}팀은 강매를 하지 않기로 선택하였습니다.')
                      break
                  elif SubsidiaryAnswer == 'O':
                      while True:
                          try: SubsidiaryHowmany = int(input(f'모회사{Teams_df.at[Teams_df.at[w,'parent'],'team name']}소유의 자사주 몇 주를 강매하시겠습니까? : '))
                          except ValueError:
                              print('숫자만 입력하십시오')
                              continue
                          SubsidiaryHowmuch = SubsidiaryHowmany * Teams_df.at[w,'price'] * roundnum/2
                          SubsidiaryOk = input(f'모회사{Teams_df.at[Teams_df.at[w,'parent'],'team name']}소유의 주식 {SubsidiaryHowmany}주를 강매하시겠습니까? 가격:{SubsidiaryHowmuch}원 O/X : ')
                          if SubsidiaryOk == 'O':
                              if Teams_df.at[w,'capital'] >= SubsidiaryHowmuch and Holdings_df.at[Teams_df.at[w,'parent'],'stock'+w] >= SubsidiaryHowmany:
                                  Teams_df.at[w,'capital'] = Teams_df.at[w,'capital'] - SubsidiaryHowmuch
                                  Teams_df.at[Teams_df.at[w,'parent'],'capital'] = Teams_df.at[Teams_df.at[w,'parent'],'capital'] + SubsidiaryHowmuch
                                  Holdings_df.at[w,'stock' + w] = Holdings_df.at[w,'stock' + w] + SubsidiaryHowmany
                                  Holdings_df.at[Teams_df.at[w,'parent'],'stock' + w] = Holdings_df.at[Teams_df.at[w,'parent'],'stock' + w] - SubsidiaryHowmany
                                  print('강매가 완료되었습니다.')
                                  IsSubsidiary()
                                  break
                              else:
                                  print('잔액이 부족하거나 모회사 보유 주식이 부족합니다.')
                          elif SubsidiaryOk == 'X':
                              print(f'자회사{Teams_df.at[w,'team name']}팀은 강매를 하지 않기로 선택하였습니다.')
                              break
                          else:
                              print('입력 잘못됨 (O 혹은 X 만 입력하시오.)')
                      break
                  else:
                      print('입력 잘못됨 (O 혹은 X 만 입력하시오.)')
```

```python
#QA 완료
def IsSubsidiary():
#자회사 해제 검사 및 수정 ----------------------------------------
    for w in Teams_df.drop(index=['S']).index:
        if Teams_df.at[w,'parent'] != 'X':
            if Holdings_df.at[Teams_df.at[w,'parent'],'stock'+ w] < 51 and w in Teams_df.at[Teams_df.at[w,'parent'], 'subsidiary']:
                Teams_df.at[Teams_df.at[w,'parent'],'subsidiary'].remove(w)
                Subsidiarys_df.at[Teams_df.at[w,'parent'],'Subsidiary'+w] = ' '
                print(f"{Teams_df.at[w,'team name']}팀은 {Teams_df.at[w,'parent']}팀의 자회사에서 해방되었습니다.")
                Teams_df.at[w,'parent'] = 'X'
                Teams_df.at[w,'parent name'] = ' '
# 자회사 여부 검사 및 수정
    for w in Holdings_df.drop(index=['S']).index: #자회사
        for t in Holdings_df.drop(index=['S']).index: #모회사
            if t == w:
                continue

            if Holdings_df.at[t,'stock'+w] >=51 and Teams_df.at[w, 'parent'] == 'X' :
                if w not in Teams_df.at[t,'subsidiary']:
                    Teams_df.at[w,'parent'] = t
                    Teams_df.at[w,'parent name'] = Teams_df.at[t,'team name']
                    Teams_df.at[t,'subsidiary'].append(w)
                    Subsidiarys_df.at[t,'Subsidiary'+w] = '자회사'

                    print(f"{Teams_df.at[w,'team name']}팀은 {Teams_df.at[t,'team name']}팀의 자회사가 되었습니다!")
    SubsidiarysReload()
```

```python
#QA 완료
def Winner2():
    #2번 승리조건 판별 함수
    for w in Teams_df.drop(index=['S']).index:
        if len(Teams_df.at[w,'subsidiary']) >= 4:
            print(f'\n\n\n\n 게임 종료! 우승팀은 {Teams_df.at[w,'team name']}팀입니다 \n\n!\n\n')
```

```python
#QA 완료
def MoneyMoney():
    # total asset 계산
    for team_id in Teams_df.drop(index=['S']).index:
        stock_asset_sum = 0
        for stock_id in Holdings_df.drop(index=['S']).index:
            # [개선된 부분] .fillna(0)를 사용하여 NaN 값을 안전하게 0으로 처리합니다.
            holding_amount = Holdings_df.at[team_id, 'stock' + stock_id]
            stock_price = Teams_df.at[stock_id, 'price']
            # NaN 값을 0으로 변환
            holding_amount = 0 if pd.isna(holding_amount) else holding_amount
            stock_price = 0 if pd.isna(stock_price) else stock_price
            stock_asset_sum += holding_amount * stock_price
        # 3. 총자산을 계산하여 DataFrame에 업데이트합니다.
        Teams_df.at[team_id, 'total asset'] = stock_asset_sum + Teams_df.at[team_id, 'capital']
    # team rank 계산
    Teams_df['team rank'] = Teams_df['total asset'].rank(method='min', ascending=False)
    Teams_df['team rank'] = Teams_df['team rank'].astype(int)
    TeamsReload()
    HoldingsReload()
```

```python
#QA 완료
def ImportTradeOrder():
    global OrderList
    global LastOrder
    global roundnum
    while True:
        TradeOrder_df = pd.read_csv('../V2/data/BP_TradeOrder.csv')
        if TradeOrder_df.empty == True:
            time.sleep(7)
            continue
        else:
            break
    if TradeOrder_df.index[-1] < LastOrder:
        time.sleep(7)

    while TradeOrder_df.index[-1] >= LastOrder:
        print(f'주문 인덱스 : {LastOrder}')
        OrderList = TradeOrder_df.iloc[LastOrder]
        LastOrder += 1
        if OrderList[1:].isnull().any():
            print('빈 주문입니다. 이 주문을 건너뜁니다.')
            continue
        OrderList = OrderList.tolist()
        OrderList.pop(0)
        try: OrderList[3] = int(float(OrderList[3]))
        except (ValueError, TypeError): # 숫자가 아니거나, NaN/None일 경우 처리
            print('주문 수량이 숫자가 아닙니다. 이 주문을 건너뜁니다.')
            continue
        if OrderList[3] <= 0:
            print('주문 수량이 0보다 작습니다. 이 주문을 건너뜁니다.')
            continue
        #회사 이름, 구매 판매 변환


        #'''!!! 보안키 !!! '''
        #'''!!! 보안키 !!! '''
        #'''!!! 보안키 !!! '''

        SecretKey = {
        "1588": "OpenAI",
        "2424": "Tesla",
        "3693": "삼성전자",
        "4885": "Palantir",
        "5959": "Instagram",
        "6256": "Amazon",
        "7749": "KakaoPay",
        "8881": "NAVER"
        }
        Teams_map = {
        "OpenAI": "A",
        "Tesla": "B",
        "삼성전자": "C",
        "Palantir": "D",
        "Instagram": "E",
        "Amazon": "F",
        "KakaoPay": "G",
        "NAVER": "H"
        }
        BuySell_map = {
        "매수": "Buy",
        "매도": "Sell"
        }
        #보안키 체크
        try:OrderList[0] = SecretKey.get(str(OrderList[0]),'X')
        except (ValueError, TypeError):
            print('보안키가 잘못됐습니다. V,T')
            continue
        except IndexError:
            print('보안키가 잘못됐습니다. I')
            continue
        if OrderList[0] == 'X':
            print('보안 키가 올바르지 않습니다.')
            continue
        print(OrderList)
        OrderList[0] = Teams_map.get(OrderList[0],'X')
        OrderList[2] = Teams_map.get(OrderList[2],'X')
        OrderList[1] = BuySell_map.get(OrderList[1],'X')
        if OrderList[0]=='X' or OrderList[1]== 'X' or OrderList[2] == 'X':
            print('주문이 올바르지 않습니다.(매핑 오류)')
            continue
        if OrderList==['A','Buy','A',20061226]:
            break
        # 주식 주문의 조건 검사 (주식을 살 돈과 매물이 있는지, 주식을 팔 매물이 있는지)
        if OrderList[1] == 'Buy':
            #자사주 1,2라운드에서 10주 이상 구매 금지
            if (roundnum == 1 or roundnum == 2) and OrderList[0]== OrderList[2] and NoNoMyStock[OrderList[0]]+OrderList[3] >= 10:
                print('1,2라운드에서 자사주 10주 이상 구매는 금지되어있습니다.')
                continue
            if (Teams_df.loc[OrderList[0],'capital'] >= int(OrderList[3]) * Teams_df.loc[OrderList[2],'price']):
                pass
            else:
                print('매수하는 데 필요한 비용보다 잔고가 부족합니다. 주문이 처리되지 않았습니다.')
                continue

            if (Holdings_df.loc['S','stock' + OrderList[2]] < int(OrderList[3])):
                print('시스템 매물이 매수하시려는 양보다 부족합니다. 주문이 처리되지 않았습니다.')
                continue
            else:
                if OrderList[0] == OrderList[2]:
                    NoNoMyStock[OrderList[0]] = NoNoMyStock[OrderList[0]]+OrderList[3]

        elif OrderList[1] == 'Sell':
            if (Holdings_df.loc[OrderList[0],'stock' + OrderList[2]] >= int(OrderList[3])):
                if OrderList[0] == OrderList[2]:
                    NoNoMyStock[OrderList[0]] = NoNoMyStock[OrderList[0]]-OrderList[3]
            else:
                print('매도하시려는 양이 보유량보다 부족합니다. 주문이 처리되지 않았습니다.')
                continue
      # 주식 주문 데이터 프레임에 실제 반영
        if OrderList[1] == 'Buy':
            Teams_df.loc[OrderList[0],'capital'] = Teams_df.loc[OrderList[0],'capital'] - int(OrderList[3]) * Teams_df.loc[OrderList[2],'price']
            Holdings_df.loc[OrderList[0],'stock' + OrderList[2]] = Holdings_df.loc[OrderList[0],'stock' + OrderList[2]] + int(OrderList[3])
            Holdings_df.loc['S','stock' + OrderList[2]] = Holdings_df.loc['S','stock' + OrderList[2]] - int(OrderList[3])
        elif OrderList[1] == 'Sell':
            Teams_df.loc[OrderList[0],'capital'] = Teams_df.loc[OrderList[0],'capital'] + int(OrderList[3]) * Teams_df.loc[OrderList[2],'price']
            Holdings_df.loc[OrderList[0],'stock' + OrderList[2]] = Holdings_df.loc[OrderList[0],'stock' + OrderList[2]] - int(OrderList[3])
            Holdings_df.loc['S','stock' + OrderList[2]] = Holdings_df.loc['S','stock' + OrderList[2]] + int(OrderList[3])
        IsSubsidiary()
        Winner2()
```

```python
#QA 완료
def RunBigGame():
    global OrderList
    global roundnum
    global NoNoMyStock
    print(f'\n{roundnum}차 빅 게임 시작!\n')
    OrderList = ['','','','']
    NoNoMyStock ={
        'A':0,
        'B':0,
        'C':0,
        'D':0,
        'E':0,
        'F':0,
        'G':0,
        'H':0
    }
    # 주문 반복 while문
    while True:
        MoneyMoney()
        # 리스트 자동 입력 (구글 폼)
        ImportTradeOrder()
        # 빅게임 종료 여부 검사
        # 빅게임 종료 코드: A,Buy,A,20061226
        if OrderList==['A','Buy','A',20061226]:
            OrderList = ['','','','']
            print('\n빅게임이 종료되었습니다.\n')
            break
        '''
        # 리스트 유효성 검사
        if (
            type(OrderList[3]) == int
        ):
            print("입력 성공:", OrderList)
        else:
            print("형식이 잘못됐습니다. 주문이 처리되지 않았습니다.")
            continue
        '''
    '''
    # TradeOrder 시트 초기화
    '''
    ForcedTrade()
    IsSubsidiary()

    Teams_df['price before'] = Teams_df['price']
    Teams_df['price'] = Teams_df['price'] + (Teams_df['capital'] / 100).astype(int)
    Teams_df['market capital'] = Teams_df['price'] * 100

    if roundnum != 3:
        Teams_df['price delta'] = Teams_df['price'] - Teams_df['price before']
        Teams_df['price ROR'] = 0.0
        mask = Teams_df['price before'] != 0
        Teams_df.loc[mask, 'price ROR'] = round(
            (Teams_df.loc[mask, 'price delta'] / Teams_df.loc[mask, 'price before'] * 100),1)

    MoneyMoney()
```
